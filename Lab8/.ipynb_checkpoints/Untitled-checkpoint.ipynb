```python
import pandas as pd
iris = pd.read_csv('archive (1)/Iris.csv')
print(iris.head())
```

```python
import seaborn as sns
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(iris, x='SepalLengthCm', hue='Species', kde=True)
plt.subplot(2, 2, 2)
sns.histplot(iris, x='SepalWidthCm', hue='Species', kde=True)
plt.subplot(2, 2, 3)
sns.histplot(iris, x='PetalLengthCm', hue='Species', kde=True)
plt.subplot(2, 2, 4)
sns.histplot(iris, x='PetalWidthCm', hue='Species', kde=True)
plt.tight_layout()
plt.show()
```

```python
# 箱线图
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=iris)
plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidthCm', data=iris)
plt.tight_layout()
plt.show()
```

```python
# 散点图矩阵
sns.pairplot(iris, hue='Species')
plt.show()
```

```python
# 小提琴图
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.violinplot(x='Species', y='SepalLengthCm', data=iris)
plt.subplot(2, 2, 2)
sns.violinplot(x='Species', y='SepalWidthCm', data=iris)
plt.subplot(2, 2, 3)
sns.violinplot(x='Species', y='PetalLengthCm', data=iris)
plt.subplot(2, 2, 4)
sns.violinplot(x='Species', y='PetalWidthCm', data=iris)
plt.tight_layout()
plt.show()
```
